```python
data_train_source = pd.read_csv('light_training_source_labels.csv')
data_train_target = pd.read_csv('light_training_target_labels.csv')
data_test_source = pd.read_csv('light_test_source_labels.csv')
```

```python
df_tmp = pd.DataFrame()
df_tmp['user_id'] = data_test_source['user_id']
data_train_source = data_train_source.drop(columns = 'user_id')
data_train_target = data_train_target.drop(columns = 'user_id')
data_test_source = data_test_source.drop(columns = 'user_id')
```

```python
train_X = data_train_source.to_numpy()
train_Y = data_train_target.to_numpy()
test_X = data_test_source.to_numpy()

train_X = np.reshape(train_X, (32000, 37, 28))
train_Y = np.reshape(train_Y, (32000, 28))
test_X = np.reshape(test_X, (8000, 37, 28))
```

```python
x = sequence_input = Input(shape=(37, 28)) # 37 wk x feature_num

x = LSTM(48, dropout = 0.375)(x)
x = Dense(128, activation='sigmoid')(x)
preds = Dense(28, activation='sigmoid')(x)

model = Model(sequence_input, preds)

model.compile(loss='binary_crossentropy',
            optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
            metrics=['acc'])
```

```python
model.fit(train_X, train_Y, epochs=10, shuffle = False)
```

```python
test_pred = model.predict(test_X)
test_df = pd.DataFrame(test_pred)
```
